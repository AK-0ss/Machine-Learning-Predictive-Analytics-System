```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor

def evaluate_model(y_test, y_pred):
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    return rmse, r2

results = {
    "Linear Regression": evaluate_model(y_test, pred_lr),
    "Random Forest": evaluate_model(y_test, pred_rf),
    "Gradient Boosting": evaluate_model(y_test, pred_gbr),
    "XGBoost": evaluate_model(y_test, pred_xgbr)
}

# Find the model with the maximum R2 score
best_r2 = -1
best_model_name = ""
model_objects = {
    "Linear Regression": lr,
    "Random Forest": rf,
    "Gradient Boosting": gbr,
    "XGBoost": xgbr # Corrected from xvgr to xgbr to match latest variable name
}

for model_name, (rmse, r2) in results.items():
    if r2 > best_r2:
        best_r2 = r2
        best_model_name = model_name

best_model = model_objects[best_model_name]

print(f"The best performing model is {best_model_name} with an R2 score of {best_r2:.4f}")
print("Assigned best_model to the corresponding model object.")
results
```

```python
import pandas as pd # Import pandas here to ensure df is always fresh
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression # Included as per previous user request

# 1. Reload original df and drop initial columns to ensure clean state
df = pd.read_csv("train.csv")
df.drop(['Alley', 'PoolQC', 'Fence', 'MiscFeature'], axis=1, inplace=True, errors='ignore')

# 2. Separate target column (unscaled) and features (unscaled)
y = df["SalePrice"]
X = df.drop("SalePrice", axis=1)

# 3. Apply one-hot encoding to X (must be done before numeric imputation for X)
X = pd.get_dummies(X, drop_first=True)

# 4. Identify numeric columns for X (after get_dummies, as new numeric cols might appear)
num_cols_X = X.select_dtypes(include=['int64', 'float64']).columns

# 5. Impute missing values in numeric columns of X with the mean
for col in num_cols_X:
    if X[col].isnull().any():
        X[col] = X[col].fillna(X[col].mean())

# 6. Create and fit a scaler for X features
X_scaler = StandardScaler()
X[num_cols_X] = X_scaler.fit_transform(X[num_cols_X])

# 7. Create and fit a separate scaler for the target variable y (unscaled)
y_scaler = StandardScaler()
# Reshape y to a 2D array as fit_transform expects it, then flatten back
y = y_scaler.fit_transform(y.values.reshape(-1, 1)).flatten()
```
